```python
import os
os.environ["OPENCV_IO_ENABLE_OPENEXR"]="1"
import cv2
import numpy as np
import glob
```

```python
def drago_tone_mapping(hdr_image,gamma=1.0,staturation=0.60,bias=0.90):
    tonemap_drago=cv2.createTonemapDrago(gamma,staturation,bias)
    ldr_image=tonemap_drago.process(hdr_image)
    ldr_image_8_bit=np.clip(ldr_image*255,0,255).astype('uint8')
    return ldr_image_8_bit
files=glob.glob('D:/HDR dataset/HDR4RTT Database/HDR4RTT Database/subset/**/*.exr',recursive=True)
for file in files:
    hdr_image=cv2.imread(file,cv2.IMREAD_ANYDEPTH | cv2.IMREAD_ANYCOLOR)
    ldr_image=drago_tone_mapping(hdr_image)
    file_name=file.split('\\')[1].split('.')[0]
    file_path='D:/HDR dataset/HDR4RTT Database/HDR4RTT Database/ldr/'+file_name+'.png'
    print(file_path)
    cv2.imwrite(file_path,ldr_image)
   
```
